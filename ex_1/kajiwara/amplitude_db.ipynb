```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
wave_data, sr = librosa.load('./sounds/sample.wav')
```

```python
cv = librosa.stft(wave_data)
```

```python
librosa.display.specshow(np.abs(cv), sr=sr, y_axis='log')
plt.savefig('amplitude-log.png')
```

```python
librosa.display.specshow(np.abs(cv), sr=sr)
plt.savefig('amplitude.png')
```

```python
db = librosa.amplitude_to_db(np.abs(cv))
```

```python
librosa.display.specshow(db)
plt.savefig('db.png')
```

```python
librosa.display.specshow(np.log10(np.abs(cv)) * 20)
```

```python
plt.imshow(np.log10(np.abs(cv)) * 20, extent=[0, len(wave_data)/sr, 0, sr/2], aspect='auto')
```
